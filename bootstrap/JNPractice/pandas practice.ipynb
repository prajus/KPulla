```python
import pandas as pd
import os
location = '/Users/santhoshvasudevan/Documents/WebDev/djangodash/AMFI/'
#directory = os.fsencode(location)
Funds_df_list = []

def getSummary(listofframes):
    x = 1
    TI,TV,Summary = pd.DataFrame(),pd.DataFrame(),pd.DataFrame()
    for chucko in listofframes:
        TI['TI_'+str(x)] = chucko['Total Investment']
        TV['TV_'+str(x)] = chucko['Total Value']
        x += 1
    
    TI.fillna(0,inplace = True)
    TV.fillna(0,inplace = True)
    Summary = pd.DataFrame(TI.sum(axis =1)).copy()
    Summary.columns = ['Overall Invested']
    Summary['Portfolio Value'] = pd.DataFrame(TV.sum(axis=1))
    return Summary


for files in os.listdir(location):
    #filename = os.fsdecode(files)
    if files.endswith(".csv"):
        df = pd.read_csv(os.path.join(location,files))
        df.set_index('Date',inplace = True)
        Funds_df_list.append(df)
        

#summary_df = getSummary(Funds_df_list)
#summary_df
```
